```python
hair_path="./Problem_C_Data/hair_dryer.tsv"
microwave_path="./Problem_C_Data/microwave.tsv"
pacifier_path="./Problem_C_Data/pacifier.tsv"

import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")

# hairdryer_info=pd.read_csv(hair_path, delimiter='\t', encoding='utf-8', index_col='review_date', parse_dates=True)

microwave_info=pd.read_csv(microwave_path, delimiter='\t', encoding='utf-8', index_col='review_date', parse_dates=True)
pacifier_info=pd.read_csv(pacifier_path, delimiter='\t', encoding='utf-8', index_col='review_date', parse_dates=True)
```

```python
m_groupby_time = microwave_info.groupby(['review_date']).star_rating.mean()

temp=m_groupby_time.head(5)
temp
```

```python
p_groupby_time = pacifier_info.groupby(['review_date']).star_rating.mean()

temp=p_groupby_time.head(5)
temp
```

```python
m_time_rating = m_groupby_time.reset_index()
m_set_index_date = m_time_rating.set_index('review_date')
m_groupby_month = m_set_index_date.groupby(pd.Grouper(freq='M')).mean()
```

```python
p_time_rating = p_groupby_time.reset_index()
p_set_index_date = p_time_rating.set_index('review_date')
p_groupby_month = p_set_index_date.groupby(pd.Grouper(freq='M')).mean()
```

```python
m_groupby_month=m_groupby_month.dropna()
p_groupby_month=p_groupby_month.dropna()
```

```python
fig_m, ax_m = plt.subplots(figsize = (20,6))    
fig_m = sns.barplot(x = m_groupby_month.index, y = m_groupby_month['star_rating'], ax=ax_m)
fig_m.tick_params(labelsize=7)
x_dates = m_groupby_month.index.strftime('%Y-%m').sort_values().unique()
ax_m.set_xticklabels(labels=x_dates, rotation=45, ha='right')
```

```python
fig_p, ax_p = plt.subplots(figsize = (20,6))    
fig_p = sns.barplot(x = p_groupby_month.index, y = p_groupby_month['star_rating'], ax=ax_p)
fig_p.tick_params(labelsize=7)
x_dates = p_groupby_month.index.strftime('%Y-%m').sort_values().unique()
ax_p.set_xticklabels(labels=x_dates, rotation=45, ha='right')
```
